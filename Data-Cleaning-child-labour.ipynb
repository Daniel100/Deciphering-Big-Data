# Data Cleansing - Performed on the child labor dataset.

This dataset was provided by [UNICEF](https://data.unicef.org) and the data can be found [here](https://data.unicef.org/resources/dataset/percentage-children-aged-5-14-years-engaged-child-labour-sex-place-residence-household-wealth-quintile/). It contains the percentage of children from 5-17 years engaged in child labout by countries and gender.


```python
# packages
import pandas as pd
import numpy as np
import openpyxl
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick                # ticker for % sign
        
# setting and magic function 
csfont = {'fontname':'Arial'} 
%matplotlib inline  
```

```python
# data load
path = "/Users/danielbosch/Desktop/Unit 9 Child Labour/child_labour.xlsx"
df = pd.read_excel(path, sheet_name = "Child labour", usecols = "A:B,D,F", skiprows = 10, nrows = 202)
```

```python
# rename columns
df.rename(
    columns = {
        "Unnamed: 0": "Countries",
        "Unnamed: 1": "Total",
        "Unnamed: 3": "Male",
        "Unnamed: 5": "Female"
    },
    inplace = True
)
# missing values
df = df.replace('-', 0)
df.head()
```

```python
# highest values
df1 = df.sort_values(by = ["Total"], ascending = False).head(10)

# horizontal barplot
custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)

ax = sns.barplot(x = df1["Total"], y = df1["Countries"], orient = "h")
ax.set_title("Highest levels of child labor per country [in %]")
ax.set_xlabel("Portion of child labor per country [in %]")
```
